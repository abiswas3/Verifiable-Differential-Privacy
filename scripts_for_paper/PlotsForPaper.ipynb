```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set('poster')
```

```python
# Secret Sharing : Varying M
d = {256:[1, 74], 512:[2, 149], 1024:[5, 300], 2048:[9, 596], 4096:[18, 1194], 8192:[36, 2390], 16384:[76, 4799], 32768:[152, 9540], 65536:[312, 19135], 131072:[656, 38717], 262144:[1279, 77326], 524288:[2415, 153082]}
df = pd.DataFrame(d, index=['Prio', 'Ours']).T

ax = df['Prio'].plot(figsize=(10,6), style='.--')
ax.set_ylabel('Milliseconds', fontsize=18);
ax.set_xlabel('Number of dimensions', fontsize=18);

df['Ours'].plot(ax=ax, figsize=(10,6), style='.-')
ax.set_title("Secret sharing timing comparison", fontsize=19)
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')

plt.savefig('/Users/ari/Desktop/secret_sharing.png')
```
