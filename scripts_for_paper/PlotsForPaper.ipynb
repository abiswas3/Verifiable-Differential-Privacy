```python
delta = 10**-10
EPSILONS = np.linspace(0.001, 2, 20)
print(EPSILONS)
s = pd.Series({epsilon: int((100/epsilon**2)*(np.log(2/delta))) for epsilon in EPSILONS})
ax = s.plot(figsize=(8,5), style=['o-','o-','*-'], color='r')
ax.set_title('Epsilon vs Number Of Coins For Fixed Delta')
ax.set_xlabel('Epsilon')
ax.set_ylabel('Number Of Coins')
ax.set_yscale('log')
```
