```python
%matplotlib inline
import pandas as pd
```

```python
output = {'coms': [], 'no_coms': []}
for key in output:
    with open("../{}.vtk".format(key)) as fp:
        output[key] = [int(x.strip()) for x in fp.readlines()[0].split(',') if len(x.strip())]
    
    
df = pd.DataFrame(output)
print('Time in microseconds')
print(df.mean())

ax = df.plot(kind='hist', figsize=(10,6), bins=10)
ax.set_xlabel('Micro seconds', fontsize=18);
ax.set_title('Histogram of runtimes over {} runs for secret sharing'.format(len(df['no_coms'])), fontsize=18);
```

```python
output = {'coms': [], 'no_coms': []}
for key in output:
    with open("../{}_input_verification.vtk".format(key)) as fp:
        output[key] = [int(x.strip()) for x in fp.readlines()[0].split(',') if len(x.strip())]
    
    
df = pd.DataFrame(output)
print('Time in milliseconds')
print(df.mean())

ax = df.plot(kind='hist', figsize=(10,6), bins=10)
ax.set_xlabel('Milli seconds', fontsize=18);
ax.set_title('Histogram of runtimes over {} runs for input verification'.format(len(df['no_coms'])), fontsize=18);
```
