```python
%matplotlib inline
import logging
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import matplotlib as mpl
import seaborn as sns
from collections import Counter
sns.set('poster')
logger = logging.getLogger()
```

```python
from accuracy_plots import make_zipf_data, precision_k, base_experiment, our_experiment
```

```python
logger.setLevel(logging.WARNING)
```

# Zipf Distribution with parameter 1.5

```python
n = 10**6 # Number of clients
tol = 10**4 # M 
k = 30 # Get TOP k most frequent items
NUM_TRIALS = 10 # Metrics reported over a average of how many trials
s = make_zipf_data(n, tol=tol, parameter=1.5)
```

```python
plotting_data = {str(i):{} for i in range(1, 5)}
EPSILONS = np.linspace(0.001, 0.1, 6)

key = str(1)
for eps in EPSILONS:
    delta = 10**-6
    if eps not in plotting_data[key]:
        plotting_data[key][eps] = []
    
    for _ in range(NUM_TRIALS):
        s_hat = base_experiment(s, eps, delta)
        s_true = s.sort_values(ascending=False)  
        p_at_k = precision_k(s_true, s_hat, k)
        plotting_data[key][eps].append(p_at_k)
    
for repeats in range(2, 5):
    key = str(repeats)
    for eps in EPSILONS:
        delta = 10**-6
        if eps not in plotting_data[key]:
            plotting_data[key][eps] = []

        for _ in range(NUM_TRIALS):
            s_hat = our_experiment(s, eps, delta, repeats)
            s_true = s.sort_values(ascending=False)  
            p_at_k = precision_k(s_true, s_hat, k)
            plotting_data[key][eps].append(p_at_k)
        
```

```python
df = pd.DataFrame(plotting_data).applymap(np.mean)
ax = df.plot(figsize=(8, 5), style=['+-','o-','.--','s:'])
ax.set_xlabel('Epsilon')
ax.set_ylabel('Avg Precision@{}'.format(k))
ax.set_xticklabels(ax.get_xticklabels(), rotation=30);
ax.set_title('Data sampled from Zipf(1.5)')
```
