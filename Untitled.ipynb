# Distribution Of Times For A Single Bit Verification And Creation

```python
fnames = ['creation', 'verification']
data = []
for name in ['{}.log'.format(i) for i in fnames]:
    with open(name) as fp:
        data.append([float(x.strip()) for x in fp.readlines()])

df = pd.DataFrame(data, index=fnames).T

means = df.mean(numeric_only=True)             #
stds = df.std(numeric_only=True)

ax = df.plot(kind='hist', figsize=(20, 6), fontsize=15)
ax.set_ylabel('Frequency', fontsize=15)
ax.set_xlabel('Time In Microseconds', fontsize=15)
ax.set_title('Distribution of Time Taken To Verify and Create a single Sigma Proof', fontsize=15);
```

# Simulation Of Latency vs Epsilon (Number Of Private Bits)

In these experiments we simulate running a single proof by drawing a random value from gaussian distribution with mean and standard deviation set to the empirical values from the above run. Then we run the trials sequentially. You can get a large speedup if we parallelise this operation.

```python
# SIMULATON OF HOW SEQUENTIAL 8
number_of_private_coins = [2**i for i in range(8, 22)]
def time_taken(n, delta, key):
    return sum(np.random.normal(means[key],
                                stds[key],
                                size=n)/10**6)

def n_to_eps(n, delta=10**-10):
    return np.sqrt(100/n*np.log(2/delta))
```

```python
s = pd.Series({n_to_eps(n, 10**-10) : time_taken(n, 10**-10, 'creation') for n in number_of_private_coins})
ax = s.plot(marker='o')
ax.set_yscale('log')
ax.set_xlabel('Epsilon')
ax.set_ylabel('Time in seconds')


s = pd.Series({n_to_eps(n, 10**-10) :time_taken(n, 10**-10, 'verification') for n in number_of_private_coins})
s.plot(marker='*', ax=ax)
ax.set_yscale('log')
ax.set_xlabel('Epsilon')
ax.set_ylabel('Time in seconds')
ax.set_title('Sigma Proof-Latency Vs Privacy Parameter')
plt.legend(df.columns)
```
